```python
training_images=training_images/255
test_images=test_images/255
```

```python
model2=tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation=tf.nn.relu),
    tf.keras.layers.Dense(10,activation=tf.nn.softmax)
])
```

```python
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy']
              )
```

```python
model2.fit(training_images,training_labels,epochs=10)
```

```python
model2.evaluate(test_images,test_labels)
```

```python
class mycallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs={}):
        if(logs.get('accuracy')>=0.95):
            print('\n Reachin accuracy 95% so Cancelling accuracy')
            self.model.stop_training=True
```

```python
callbacks=mycallback()
mnist=tf.keras.datasets.mnist
(training_image,training_labels),(test_images,test_labels)=mnist.load_data()
training_image=training_image/255
test_images=test_images/255
model3=tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation=tf.nn.relu),
    tf.keras.layers.Dense(10,activation=tf.nn.softmax)
])
model3.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy']
              )
model3.fit(training_image,training_labels,epochs=10,callbacks=[callbacks])
```
